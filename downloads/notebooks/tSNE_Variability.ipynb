# Are tSNE/viSNE Embeddings Reproducible?

On September 10, 2014 I gave a talk to the MD Anderson Flow Cytometry User Group.  One persistent question was "are viSNE runs reproducible?"  It depends on what you mean by reproducible.  If you expect broadly similar patterns in the 'eyeball norm', then yes; but if you expect identitical patterns, unfortunately, they are not.  To demonstrate, I've run the t-SNE algorithm multiple times on the same data and compared the results.  In this example, we use clinical data from an AML patient's peripheral blood 5 hours after treatment.

```python
# <!-- collapse=True -->
fig, axes = plt.subplots(2, 5, figsize=(3*5, 3*2))
for ax, fd in zip(axes.flat, fd_tenthousand):
    x = fd['tsne1']
    y = fd['tsne2']
    sns.kdeplot(x, y, cmap=cmap, shade=True, cut=2, ax=ax)
    ax.set_aspect('equal')
    ax.set_xticklabels('')
    ax.set_yticklabels('')
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xlim(-52,52)
    ax.set_ylim(-52,52)
fig.subplots_adjust(wspace=0,hspace=0)
```

The embedding that t-SNE uses is invariant under translation, rotation, and reflection; as such, we cannot expect each resulting plot to be identitical.  (Hence the importance to compare t-SNE embeddings that were computed at the same time!)  However, given two of these plots we can find the rotation and reflection of one to best match another by solving the [Orthgonal Procrustes Problem](https://en.wikipedia.org/wiki/Orthogonal_Procrustes_problem).  After this modification, the plots look similar.

```python
# <!-- collapse=True -->
fig, axes = plt.subplots(2, 5, figsize=(3*5, 3*2))
X = fd_tenthousand[0].matrix('tsne1','tsne2')
for ax, fd in zip(axes.flat, fd_tenthousand):
    Y = fd.matrix('tsne1','tsne2')
    A = inner(X.T,Y.T)
    U,S,V = svd(A)
    R = dot(U,V)
    Z = dot(R,Y.T).T
    sns.kdeplot(Z[:,0], Z[:,1], cmap=cmap, shade=True, cut=2, ax=ax)
    ax.set_aspect('equal')
    ax.set_xticklabels('')
    ax.set_yticklabels('')
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xlim(-52,52)
    ax.set_ylim(-52,52)
fig.subplots_adjust(wspace=0,hspace=0)
```

Visually, these plots are very similar with a few exceptions: the island moves around; the tip of the peninsula bends; and the central bay in the first, becomes a lake in the fifth.  Comparing the proximity of points in each embedding above to every other embedding, we get the following matrix of similarites below.  Each solution varies by about 20%.

```python
# <!-- collapse=True -->
err = np.zeros((10,10))
for j, fd1 in enumerate(fd_tenthousand):
    for k, fd2 in enumerate(fd_tenthousand):
        if j != k:
            X = fd1.matrix('tsne1','tsne2')
            Y = fd2.matrix('tsne1','tsne2')
            A = inner(X.T,Y.T)
            U,S,V = svd(A)
            R = dot(U,V)
            Z = dot(R,Y.T).T
            E = X-Z
            err[j,k] = 100*np.sum(np.sqrt(np.sum(E**2,axis=1)))/X.shape[0]/np.mean(
                    np.sqrt(np.sum(X**2,axis=1))+np.sqrt(np.sum(Y**2,axis=1)))
plt.imshow(err.T, interpolation="nearest", cmap = cmap, extent = [0.5,10.5,0.5,10.5],vmax=100,origin='lower')
plt.xticks(range(1,11))
plt.yticks(range(1,11))
plt.colorbar();
plt.title('Mean Error of the Approximation (precent)')
```

## Looking at All the Data

By default in both [SightOf](http://www.c2b2.columbia.edu/danapeerlab/html/cyt.html) (aka viSNE) and [flowML](http://flowml.mdanderson.edu:5000/) take a random sample of points to reduce computational complexity; this introduces an additional degree of randomness.  However, even with the same points, if we compute the t-SNE embedding starting from a different random initial guess, the solution will be different.  This is because there are many [local minima](https://courses.cs.washington.edu/courses/csep521/99sp/lectures/lecture06/sld014.htm) which the algorithm could converge to.  Finding the global minima is for nonconvex problems like t-SNE is an unsolved problem in mathematics and one for which we do not expect a solution. 

```python
# <!-- collapse=True -->
fig, axes = plt.subplots(2, 5, figsize=(3*5, 3*2))
X = fd_full[0].matrix('tsne1','tsne2')
for ax, fd in zip(axes.flat, fd_full):
    Y = fd.matrix('tsne1','tsne2')
    A = inner(X.T,Y.T)
    U,S,V = svd(A)
    R = dot(U,V)
    Z = dot(R,Y.T).T
    sns.kdeplot(Z[:,0], Z[:,1], cmap=cmap, shade=True, cut=2, ax=ax)
    ax.set_aspect('equal')
    ax.set_xticklabels('')
    ax.set_yticklabels('')
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xlim(-40,40)
    ax.set_ylim(-40,40)
fig.subplots_adjust(wspace=0,hspace=0)
```

(N.B.: van der Maaten's [code](http://homepage.tudelft.nl/19j49/t-SNE.html) initializes from the same random start on each run.  For the examples above, I have modified it to run with a different random seed.)

Unlike with the random sampling, runs on the full data create dramatically different results.  As shown below, run As we can see below, runs on the full data produce surprisingly different results: run 5 (in the upper right above) is nothing like any of the other runs.  However some runs are very similar: e.g., 4, 8, and 9.  The number of iterations were fixed with this experiment, but I suspect with additional iterations allowed, the results would be more similar.

```python
# <!-- collapse=True -->
err = np.zeros((10,10))
for j, fd1 in enumerate(fd_full):
    for k, fd2 in enumerate(fd_full):
        if j != k:
            X = fd1.matrix('tsne1','tsne2')
            Y = fd2.matrix('tsne1','tsne2')
            A = inner(X.T,Y.T)
            U,S,V = svd(A)
            R = dot(U,V)
            Z = dot(R,Y.T).T
            E = X-Z
            err[j,k] = 100*np.sum(np.sqrt(np.sum(E**2,axis=1)))/(X.shape[0]*np.mean(
                    np.sqrt(np.sum(X**2,axis=1))))
plt.imshow(err, interpolation="nearest", cmap = cmap, extent = [0.5,10.5,0.5,10.5], vmax=100, origin='lower')
plt.xticks(range(1,11))
plt.yticks(range(1,11))
plt.colorbar();
plt.title('Relative Error (percent) of approximation')
```

In addition to running additional iterations, another approach to get a more 'definiative' plot would be to run multiple iterations and simply select the one with the minimum residual. However, this would be expensive because each run of the Barnes-Hut t-SNE on full datasets takes about 30 minutes.  The moral of the story is simple: although the general patterns of a particular t-SNE embedding appear in other solutions, no two embeddings are identical.  None the less, the t-SNE embedding is still an appropreate tool for visualization and gating.
