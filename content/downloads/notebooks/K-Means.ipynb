## Building the Clusters
For this example, we will generate $k$ clusters, equally spaced on a circle of radius $k$.  Each cluster will have the same number of points, $n$, and will have members generated from a normal distribution with standard deviation $\sigma$.  

```python
import numpy as np
from math import pi, sin, cos
%matplotlib inline
```

```python
k = 3 # Number of clusters we will generate
n = 100 # Number of points per cluster
sigma = 0.5 # standard deviation
data = np.empty([0,2])

for j in range(k):
    center = np.array([k*sin(2*pi*j/k),k*cos(2*pi*j/k)])
    data = np.concatenate( (data, center + sigma*np.random.randn(n,2)))
```

## Plotting the Generated Clusters
Next, we visualize the clusters using ``matplotlib`` to display an inline plot.

```python
import matplotlib.pyplot as plt
for j in range(k):
    index = j*n + np.arange(n)
    plt.plot(data[index,0],data[index,1],'.',markersize=20)
plt.axis('equal')
plt.show()
```

## K-Means
Next, we use k-means clustering algorithm on these points to attempt to recover custer membership using the version in ``scipy.cluster``.

```python
from scipy.cluster.vq import kmeans2, whiten
wdata = whiten(data)
# fix seed for random sampling
np.random.seed(0)
centroid, label = kmeans2(wdata,k)
```

```python
for j in range(k):
    index = np.argwhere(label==j)
    plt.plot(data[index,0],data[index,1],'.',markersize=20)
plt.axis('equal')
plt.show()   
```

Although k-means has ordered the clusters differently, the result is exactly what we want: each data point has been assigned to the correct cluster. 

## Exploring K-Means Interactively in an Active Notebook
As we can see, K-means has correctly assigned the clusters in the above example.  But what happens as we make the clusters larger or move them closer together?

```python
from IPython.html.widgets import interact
def sweep_sigma_kmeans(sigma):
    # Fix random number generator to make plots consistent
    np.random.seed(0)
    # Generate data
    data = np.empty([0,2])
    # True labels
    true_labels = np.empty([0])
    for j in range(k):
        center = np.array([k*sin(2*pi*j/k),k*cos(2*pi*j/k)])
        data = np.concatenate( (data, center + sigma*np.random.randn(n,2)))
        true_labels = np.concatenate( (true_labels, j*np.ones(n)))
    # Perform K-means
    wdata = whiten(data)
    centroid, label = kmeans2(wdata,k)
    
    # Build mapping between K-means and original labels (should use marriage algorithm)
    label_map = np.zeros(3)
    for j in range(k):
        local_labels = label[range(n*j,n*j+n)]
        label_map[j] = np.argmax(np.bincount(local_labels))    

    # Plot
    for j in range(k):
        # plot those that match their indices
        index = np.argwhere( (label_map[label]==j) * (true_labels==j))
        plt.plot(data[index,0],data[index,1],'.',markersize=20)
    plt.gca().set_color_cycle(None)
    for j in range(k):
        # plot those that match their indices
        index = np.argwhere( (label_map[label]==j) * (true_labels!=j))
        plt.plot(data[index,0],data[index,1],'x',markersize=5)
    plt.xlim(-8,8)
    plt.ylim(-8,8)
    plt.show()
interact(sweep_sigma_kmeans,sigma=(0,3,0.1))
```

If you are running this notebook locally, IPython's interact module will provide slider and plot that updates dynamically based on how the user moves the slider left and right.  'x' denotes a wrongly assigned element.

## Static Web Based Interactive Graphics
The above example only works with an open notebook running locally, but can we do the same on the web with static HTML?  We can, thanks to [Jake VanderPlas](http://www.astro.washington.edu/users/vanderplas/) who recently released the [``ipywidgets`` package](https://jakevdp.github.io/blog/2013/12/05/static-interactive-widgets/) that allows the results to be precomputed and then diplayed using JavaScript.

There is only a slight modification of the code used above to use this capability: the code now returns the figure we created. Everything else proceeds as before.

```python
%matplotlib inline
from ipywidgets import StaticInteract, RangeWidget, RadioWidget

def sweep_sigma_kmeans(sigma):
    # Plot setup
    fig, ax = plt.subplots(figsize=(6,6))
    ax.axis('equal')
    # Fix random number generator to make plots consistent
    np.random.seed(0)
    # Generate data
    data = np.empty([0,2])
    true_labels = np.empty([0])
    for j in range(k):
        center = np.array([k*sin(2*pi*j/k),k*cos(2*pi*j/k)])
        data = np.concatenate( (data, center + sigma*np.random.randn(n,2)))
        true_labels = np.concatenate( (true_labels, j*np.ones(n)))
    # Perform K-means
    wdata = whiten(data)
    centroid, label = kmeans2(wdata,k)
    
    # Build mapping between K-means and original labels (should use marriage algorithm)
    label_map = np.zeros(3)
    for j in range(k):
        local_labels = label[range(n*j,n*j+n)]
        label_map[j] = np.argmax(np.bincount(local_labels))    

    
    # Plot
    for j in range(k):
        index = np.argwhere( (label_map[label]==j) * (true_labels==j))
        ax.plot(data[index,0],data[index,1],'.',markersize=20)
    ax.set_color_cycle(None)
    for j in range(k):
        # plot those that match their indices
        index = np.argwhere( (label_map[label]==j) * (true_labels!=j))
        plt.plot(data[index,0],data[index,1],'x',markersize=5)
    ax.set_xlim(-8,8)
    ax.set_ylim(-8,8)
    return fig

StaticInteract(sweep_sigma_kmeans,sigma=RangeWidget(0,3,0.3))
```

An interesting feature that emerges with this interactive demo is that k-means fails to find three clusters.  With random starting values, it can fall into a local minima and fail to resolve one of the three clusters.  Hence the typical advice to run K-Means multiple times to prevent this failure.
