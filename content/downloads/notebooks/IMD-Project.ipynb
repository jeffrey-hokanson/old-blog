#Extracting Better Data from the CyTOF Instrument

<small> This is a proposal for a senior design project at Rice Univeristy </small>

The [CyTOF](http://www.dvssciences.com/cytof-instrument.php) is a new instrument for measuring the properties of an individual cells.  This device improves upon traditional [flow cytometry](http://en.wikipedia.org/wiki/Flow_cytometry) instruments that measure the expression of cell surface markers and internal markers by staining the cell with fluorescent tags attached to antibodies.  Instead of using fluorescent tags, the CyTOF uses heavy metal ions that allow additional tags with decreased cross-talk between different tags.  Currently, only a limited amount of information is extracted from the raw data the CyTOF collects in a primative manner.  The goal of this project is to extract additional information about each cell from this raw data and to better separate two cells that enter the device at the same time.

### Overall Goals
* Develop an algorithm to improve separation of doublets (pairs of cells in the machine at the same time).
* Extract additional cellular properties by considering the raw data.
* Improve extraction of cellular properties.
* Develop software usable by the flow cytometry community.

## Typical Data
The following example shows typical data that comes off of the CyTOF.  Each marker is one row in the plot below, and the quantity of the marker show in the vertical axis.  Each cell is stained with Ir(191) and Ir(193) to separate cells from noise.  These cells appear over a finite time, as when the cell is vaporized in the processes of determining which heavy metal tags it has.  Cisplantin and dead channels report cells that are dead.  Most of the time, cells appear as broad, Gaussian-shaped humps as seen on the right.  However, more complicated structures can appear, such as those on the left.  Is this a single cell, or two cells close together?  The current software cannot distinguish these two cases.


```python
# <!-- collapse=True -->
import imd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline
# Prettier plotting options
try:
    from mpltools import style
    from mpltools import layout
    style.use('ggplot')
except:
    pass
import numpy as np


def stream_plot(data, start, step = 1000):
    """ Plotting function for IMD files"""
    ncol = data.ncol
    fig = plt.figure(figsize=(16, 16))
    gs = gridspec.GridSpec(ncol,1, wspace = 5.0, hspace = 0.0)
    ax = []
    time = np.arange(start, start+step)
    x = data[start:start+step]
    for j in range(ncol):
        ax.append(plt.subplot(gs[j,0]))
        ax[j].plot(time, x[:,j])
        ax[j].set_yticklabels([],visible = False)
        if j != ncol - 1:
            ax[j].set_xticklabels([], visible = False)
        ax[j].set_ylabel(data.tags[j])
        if divmod(j,2)[1] == 0:
            ax[j].yaxis.labelpad = 25
        else:
            ax[j].yaxis.labelpad = 5
```

```python
data = imd.read('test.imd')
stream_plot(data, 118000, 1200)
```
