The t-SNE embedding requires the optimization of the following function:
$$
C(\lbrace y_i \rbrace_i) = \sum_{i\ne j} p_{i,j} \log \frac{p_{i,j}}{q_{i,j}}
$$
where
\begin{align}
 p_{j|i} &= \frac{ e^{-d(x_i,x_j)^2/(2\sigma_i^2)}}{\sum_{k\ne i} e^{-d(x_i,x_k)^2/(2\sigma_i^2)}} \qquad p_{i,i} = 0 \\
 p_{i,j} &= \frac{ p_{j|i} + p_{i|j}}{2N} \\
q_{i,j} &= \frac{ (1+ \| y_i - y_j\|^2)^{-1}}{\sum_{k\ne \ell} (1+ \| y_k - y_\ell\|^2)^{-1}}
\end{align}

The gradient is then 
\begin{align}
\frac{\partial C}{\partial y_i} = 4\sum_{j\ne i} (p_{i,j} - q_{i,j}) q_{i,j} Z(y_i - y_j)
\end{align}
where $Z = \sum_{k\ne \ell} (1+ \| y_k - y_\ell\|^2)^{-1}$.

There is one simplification we can make, which is that
\begin{align}
    q_{i,j} Z = (1 + \| y_i - y_j\|^2)
\end{align}


## Computing $\sigma_i$
Both Hilton and Roweis and van der Maaten's paper allude to picking $\sigma_i$ to ensure the perplexity is within a particular range.  They use a binary search to find this range, but it is not clear what they are optimizing.

Examining the source code of [van der Maaten's Python implementation](http://homepage.tudelft.nl/19j49/t-SNE_files/tsne_python.zip), 

\begin{align}
    P_{i} &= e^{-d(x_i, x_j)\beta} \\
    H &= \log \sum_{i\ne j} P_{i,j} + \beta \frac{\sum_{i \ne j} d(x_i,x_j) P_{i,j}}{\sum_{i\ne j} P_{i,j}} 
\end{align}

Then we optimize $H$ to be near $\log$ of perplexity by tweaking $\beta$, with $\sigma = \sqrt{1/\beta}$.
